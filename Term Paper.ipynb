```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
def init(n):
    lst = []
    for i in range(n):
        lst.append(0)
    return lst
```

```python
m_0 = 5
m = 5
n = m_0
A = np.zeros((n,n) , dtype = int)
```

```python
def add_edge(adj, i, n):
    adj[i][n-1] = adj[i][n-1] + 1
    adj[n-1][i] = adj[n-1][i] + 1
    return adj
```

```python
for t in range(2000):
    n_prev = n
    n = n + 1
    A_bef = A
    A = np.zeros((n,n),dtype = int)
    A[0:n_prev,0:n_prev] = A_bef
    wts = init(n_prev)
    for i in range(n_prev):
        wts[i] = sum(A_bef[i])
    #print(wts)
    #print(list(range(n_prev)))
    lst = random.choices(list(range(n_prev)), weights = wts, k = 5)
    #print(lst)
    for i in lst:
        A = add_edge(A,i,n)
    print(t)
```

```python
degree = init(n)
for i in range(n):
    degree[i] = sum(A[i])
```

```python
probability = []
for i in range(max(degree) + 1):
    probability.append(degree.count(i)/2005)
```

```python
plt.loglog(list(range(len(probability))) , probability)
plt.ylim(0.01,0.30)
plt.xlim(5,100)
```
